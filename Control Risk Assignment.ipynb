```python
#import required modules
from rapidfuzz import process, fuzz
import pandas as pd
import numpy as np
```

```python
#read excel file
df = pd.read_excel(r'D:\Control Risk Interveiw\2021 Annual Report ONGC.xlsx')
df.shape
```

```python
#get duplicate records
duplicates=df.duplicated()
print("Total duplicate records:",duplicates.sum())
```

```python
#remove all the duplicate rows from the data set
df1 = df.drop_duplicates()
df1 .head(10)
```

```python
# creating loop to get fuzz.ratio and matched company name

df_x = pd.DataFrame(columns = ['Name of the Company','Fuzzy Ratio', 'Matched Company Name'])

for com_name in df1['Name of the Company']:
    ratio = []
    names = []
    n2 = []
    for com_by_name in df1['Name of the Company']:
        ratio.append(fuzz.ratio(com_name, com_by_name))
        names.append(com_by_name)
        n2.append(com_name)
    df_n = pd.DataFrame(list(zip(n2,ratio,names)), columns =['Name of the Company','Fuzzy Ratio', 'Matched Company Name'])
    df_nn = pd.DataFrame(df_n.sort_values(by=['Fuzzy Ratio']).iloc[-2]).T
    df_x = pd.concat([df_x,df_nn])

df_x.head(10)
```

```python
newdf = df1.set_index('Name of the Company').join(df_x.set_index('Name of the Company'))
newdf.head(10)
```

```python
#perform data transformation in newdf

#Rename of Column name
newdf.rename(columns={"Fuzzy Ratio":"Match"}, inplace=True)

#Check the Match
newdf['Match'] = np.where(newdf['Match']>=65, 'Yes', 'No')

#Remove element where matche is No
newdf['Matched Company Name'].mask(newdf['Match'] == "No", "", inplace=True)
newdf.head(10)
```

```python
newdf['Proportion of ownership interest as March 31, 2021'] = round(newdf['Proportion of ownership interest as March 31, 2021']*100,2)
```
